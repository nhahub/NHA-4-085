```python
from pyspark.sql.functions import col

# Read Silver data
df = spark.table("depi_project.philo_files.silver_linkedin_indeed")

# Align explicitly to gold's schema (name, order, and type)
df = df.select(
    col("title").cast("string").alias("title"),
    col("company").cast("string").alias("company"),
    col("location").cast("string").alias("location"),
    col("url").cast("string").alias("url"),
    col("source").cast("string").alias("source"),
    col("date").cast("string").alias("date"),
)

# Append to Gold Table
df.write \
    .format("delta") \
    .mode("append") \
    .saveAsTable("depi_project.philo_files.gold_jobs")
```
