```python
from pyspark.sql.functions import col

df = spark.table("depi_project.philo_files.bronze_linkedin_indeed")

df = df.select(
    col("title").cast("string").alias("title"),
    col("company").cast("string").alias("company"),
    col("location").cast("string").alias("location"),
    col("job_url").cast("string").alias("url"),
    col("source").cast("string").alias("source"),
    col("date_posted").cast("string").alias("date"),
)

df = df.dropDuplicates()
df = df.na.drop()

df.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("depi_project.philo_files.silver_linkedin_indeed")
```
